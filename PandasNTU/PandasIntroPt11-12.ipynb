```python
#Intro to Pandas found here: https://libguides.ntu.edu.sg/python/intropandas
#Parts 11 & 12
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
df = pd.read_csv('iris.csv')
```

# Data Classification and Summary:

Classifying your data for more refined and accurate understanding of the data is an important faucet of data analysis, hence, we can utilize Pandas to carry out such operations by organizing and summarizing our data.

`Groupby` Function

One way we can classify our data is using the group by function:

`df.groupby(by=grouping_columns)[columns_to_show].function()`

`Pivot_table` Function

For summarizing data, we can employ similar methods to that of the excel function, pivot tables, to avoid showing large chunks of data

`df.pivot_table([rows_to_be_displayed'],`

               `[columns_to_be_displayed], aggfunc=function)`

```python
#let's find summary statistics about each 'variety'
df.groupby(by = ['variety'])[('petal.length')].describe()
#note that our by = parameter defines which values (variety) to group by
#and that our second [] tells which column(s) to show
```

```python
#now let' try pivot_table to show avg petal length and width for each variety
df.pivot_table(['petal.length','petal.width'],
['variety'],aggfunc = 'mean')
```

# Data Visualization

Since Pandas simply helps us with data structuring, we will need to employ Pandas in conjunction with other modules to help visualize this data. Two common modules are seaborn and matplotlib. With seaborn and matplotlib we can create some visualisations quickly using the following functions

### 1. Line plots

sns.lineplot(x=X_FIEL',y=Y_FIELD,data=DATFRAME)

### 2. Regression plots

ay=sns.relplot(x=X_FIELD,y=Y_FIELD,hue=DATA_CLASSIFIER,data=DATFRAME)

### 3. Histogram Plots

plt.hist(X_FIELD, bins=NUMBER_OF_BINS)

### 4.  Pair Plots

sns.pairplot(DATAFRAME, hue=DATA_CLASSIFIER,height=HEIGHT)


## First we'll use sns.lineplot

```python
#first let's plot petal length by width to see correlation... solid!
sns.lineplot(x = 'petal.width',y = 'petal.length',data = df)
```

## Now we'll make a regression with sns.relplot


```python
#now let's use regression plot for petal width predicting sepal width
regression1 = sns.relplot(x = 'petal.width', y = 'sepal.width',hue = 'variety',data = df)

#how about petal length and width?
regression2 = sns.relplot(x = 'petal.width', y = 'petal.length',hue = 'variety',data = df)
#note that these results are just what we saw with the line graph before, expected
```

## Now we'll make a histogram


```python
#simply bin petal width
plt.hist(df['petal.width'],bins = 30)
plt.title('Iris data petal width histogram:')
plt.show()
```

## Finally a pairplot with seaborn:


```python
#essentially a correlation matrix w/ scatter plots and hist
sns.pairplot(df,hue = 'variety',height = 3)
plt.show()
```
